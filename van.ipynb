```python
import os
import torch
from PIL import Image
from sklearn.model_selection import train_test_split
import shutil
```

```python
van_based = r"E:\projects\vangoughGAN\vangough"
paprika_based = r"E:\projects\vangoughGAN\paprika"
van_resised = r"E:\projects\vangoughGAN\vangough_resized"
paprika_resised = r"E:\projects\vangoughGAN\paprika_resized"

os.makedirs(van_based, exist_ok=True)
os.makedirs(paprika_based, exist_ok=True)
os.makedirs(van_resised, exist_ok=True)
os.makedirs(paprika_resised, exist_ok=True)

def data_prepare(oslist,base_dist,target_dist):
    for im in oslist:
        img=Image.open(os.path.join(base_dist,im))
        img = img.resize((128, 128), Image.LANCZOS)
        img.save(os.path.join(target_dist, im))

data_prepare(os.listdir(van_based),van_based,van_resised)
data_prepare(os.listdir(paprika_based),paprika_based,paprika_resised)
```

```python
paprika=sorted([im for im in os.listdir(paprika_resised)])[:1200] 
paprika_train, paprika_test = train_test_split(paprika, test_size=0.1, random_state=42)
vangough=sorted([im for im in os.listdir(van_resised)])[:1200] 
vangough_train, vangough_test = train_test_split(vangough, test_size=0.1, random_state=42)

trainA_dir = r"E:\projects\vangoughGAN\pytorch-CycleGAN-and-pix2pix\datasets\pap2van\trainA"
testA_dir = r"E:\projects\vangoughGAN\pytorch-CycleGAN-and-pix2pix\datasets\pap2van\testA"
trainB_dir = r"E:\projects\vangoughGAN\pytorch-CycleGAN-and-pix2pix\datasets\pap2van\trainB"
testB_dir = r"E:\projects\vangoughGAN\pytorch-CycleGAN-and-pix2pix\datasets\pap2van\testB"

def copy_files(file_list, src_dir, dest_dir):
    os.makedirs(dest_dir, exist_ok=True)  
    for filename in file_list:
        src_path = os.path.join(src_dir, filename)
        dest_path = os.path.join(dest_dir, filename)
        shutil.copy2(src_path, dest_path)  

copy_files(paprika_train, paprika_resised, trainA_dir)
copy_files(paprika_test, paprika_resised, testA_dir)
copy_files(vangough_train, van_resised, trainB_dir)
copy_files(vangough_test, van_resised, testB_dir)
```
